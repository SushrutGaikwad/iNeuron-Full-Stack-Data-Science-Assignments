## Question
**What is the difference between `break` and `continue`?**

**Answer:**
`break` stops the loop, whereas `continue` restarts the loop without executing statements below it, i.e., without exhausting the loop.

## Question
**In a for loop, what is the difference between `range(10)`, `range(0,10)`, and `range(0,10,1)`?**

**Answer:**
They are the same.

````python
for i in range(10):
    print(i)
````

````python
for i in range(0,10):
    print(i)
````

````python
for i in range(0,10,1):
    print(i)
````

## Question
**Write a short program that prints numbers from 1 to 10 using a for loop and a while loop.**

````python
# For loop
for i in range(1,11):
    print(i)
````

````python
# While loop
i = 1
while i < 11:
    print(i)
    i+=1
````
