## Question 1
**What advantages do Excel spreadsheets have over CSV spreadsheets?**

**Answer:**

Excel spreadsheets have multiple tabs, parameters like font size, font colour, font style, etc., can be customized, cells themselves are also customizable, we can add images, shapes, plots, etc.

## Question 2
**What do you pass to `csv.reader()` and `csv.writer()` to create `reader` and `writer` objects?**

**Answer:**

We pass a `File` object.

## Question 3
**What modes do `File` objects for `reader` and `writer` objects need to be opened in?**

**Answer:**

For `reader` object, they need to be opened in the mode `'rb'`, whereas for the `writer` object they need to be opened in the mode `'wb'`.

## Question 4
**What method takes a list argument and writes it to a CSV file?**

**Answer:**

`writerow()`

## Question 5
**What do the keyword arguments `delimiter` and `lineterminator` do?**

**Answer:**

`delimiter` changes the string to separate cells in a row, whereas `lineterminator` changes the string to separate rows.
