## Question 10
**Does the following expression cause an error?**

`'I have eaten ' + 99 + ' burritos.'`

**If yes, how can you fix it?**

**Answer:**

This expression will cause an error because we cannot concatenate an integer to a string. To fix this error, we would have to convert `99` into a string. We can do this using the `str()` function, which is shown in the following cell.

```python
'I have eaten ' + str(99) + ' burritos.'
```
