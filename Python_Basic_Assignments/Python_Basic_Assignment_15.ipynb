## Question 7
**Write a generator, `genPrimes`, that returns the sequence of prime numbers on successive calls to its `next()` method: 2, 3, 5, 7, 11, ...**

```python
def getPrimes():
    prime = 1
    guess = 2
    while True:
        for i in range(2, guess):
            if guess%i == 0:
                break
        else:
            prime = guess
            yield prime
        guess += 1

primes = getPrimes()
```
