## Question 6
**If a dictionary is stored in `spam`, what is the difference between the expressions `'cat' in spam` and `'cat' in spam.values()`?**

**Answer:**

`'cat' in spam` checks whether there is a `'cat'` key in the dictionary, whereas `'cat' in spam.values()` checks whether there is a value `'cat'` for one of the keys in the dictionary.
