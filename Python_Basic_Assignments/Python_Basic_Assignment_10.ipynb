## Question 1
**How do you distinguish between `shutil.copy()` and `shutil.copytree()`?**

**Answer:**

`shutil.copy()` copies a single file, whereas `shutil.copytree()` copies an entire folder and its contents.

## Question 5
**Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.**

**Answer:**

I will copy `.png` files from a folder consisting files with multiple extensions to another folder.

```python
import os
import shutil

class SelectiveCopy:
    def __init__(self, directory_path, extension, destination_path):
        self.directory_path = directory_path
        self.extension = extension
        self.destination_path = destination_path
    def copy(self):
        os.chdir(self.directory_path)
        for foldername, subfolders, filenames in os.walk(self.directory_path):
            for filename in filenames:
                if filename.endswith(self.extension):
                    shutil.copy(os.path.join(foldername, filename), self.destination_path)
```

```python
directory_path = 'C:\iNeuron\Assignments\Assignment_10_code\Folder_1'
extension = '.png'
destination_path = 'C:\iNeuron\Assignments\Assignment_10_code\Folder_2'

copy_png = SelectiveCopy(directory_path=directory_path, extension=extension, destination_path=destination_path)
copy_png.copy()
```
