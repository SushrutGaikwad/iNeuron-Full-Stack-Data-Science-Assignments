## Question 1
**To what does a relative path refer?**

**Answer:**

Relative paths are the paths relative to the current working directory.

## Question 2
**What does an absolute path start with in your operating system?**

**Answer:**

In my operating system, absolute path starts with the drive name, e.g., `C:\...`

## Question 3
**What do the functions `os.getcwd()` and `os.chdir()` do?**

**Answer:**

`os.getcwd()` gives the current working directory, whereas `os.chdir()` changes the current working directory.

## Question 4
**What are the `.` and `..` folders?**

**Answer:**

`.` is the current folder, `..` is the parent folder.

## Question 5
**In `C:\bacon\eggs\spam.txt`, which part is the dir name, and which part is the base name?**

**Answer:**

`C:\bacon\eggs\` is the directory, whereas `spam.txt` is the base name.

## Question 6
**What are the three “mode” arguments that can be passed to the `open()` function?**

**Answer:**

The arguments `r`, `w`, and `a` can be passed for the read, write, and the append modes respectively.

## Question 7
**What happens if an existing file is opened in write mode?**

**Answer:**

The file is wiped out and overwritten.

## Question 8
**How do you tell the difference between `read()` and `readlines()`?**

**Answer:**

`read()` gives all the content of the file at once, whereas `readlines()` gives all the sentences as a list of strings.

## Question 9
**What data structure does a shelf value resemble?**
